```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/drive/MyDrive/Healthifier/BreastCancer/ML/breastCancer.csv')
```

```python
print(data.shape)
data.head()
```

```python
data.columns
```

```python
data = data.drop(columns=['id', 'Unnamed: 32'])
```

```python
data.tail()
```

```python
data['diagnosis'] = data['diagnosis'].map({'B':0, 'M':1})
```

```python
data.head()
```

```python
X = data.drop(columns=['diagnosis'])
y = data['diagnosis']
```

```python
X[:1].values
```

##**Scaling the Dataset**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
standardScaler = StandardScaler()
```

```python
X_Scaled = standardScaler.fit_transform(X)
```

##**Splitting the Data**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_Scaled, y, test_size=0.2, random_state=1)
```

##**Models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
knn = KNeighborsClassifier(n_neighbors = 5)
svmLinear = SVC(kernel = 'linear', random_state=0)
svmRBF = SVC(kernel = 'rbf', random_state=0)
gnb = GaussianNB()
decisionTree = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
randomForest = RandomForestClassifier(n_estimators = 10, criterion='entropy', random_state=0)
```

##**Training and Testing**

```python
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
```

```python
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
```

```python
svmLinear.fit(x_train, y_train)
y_pred_svmLinear = svmLinear.predict(x_test)
```

```python
svmRBF.fit(x_train, y_train)
y_pred_svmRBF = svmRBF.predict(x_test)
```

```python
gnb.fit(x_train, y_train)
y_pred_gnb = gnb.predict(x_test)
```

```python
decisionTree.fit(x_train, y_train)
y_pred_decisionTree = decisionTree.predict(x_test)
```

```python
randomForest.fit(x_train, y_train)
y_pred_randomForest = randomForest.predict(x_test)
```

##**Accuracy**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

```python
print(f'Logistic Regression: {accuracy_score(y_test, y_pred_lr)}')
print(f'K Nearest Neighbors: {accuracy_score(y_test, y_pred_knn)}')
print(f'SVM Linear: {accuracy_score(y_test, y_pred_svmLinear)}')
print(f'SVM RBF: {accuracy_score(y_test, y_pred_svmRBF)}')
print(f'Gaussian Naive Bayes: {accuracy_score(y_test, y_pred_gnb)}')
print(f'Decision Tree: {accuracy_score(y_test, y_pred_decisionTree)}')
print(f'Random Forest: {accuracy_score(y_test, y_pred_randomForest)}')
```
